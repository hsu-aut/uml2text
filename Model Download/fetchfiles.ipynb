Download a single file 


```python
import requests

url = "https://raw.githubusercontent.com/09421/Track-A-Bus/master/1%20-%20Systemdesign/3%20-%204th/04_Systemdesign/Diagrammer/SSD/UC1.2%20Simple%20Sequence%20Diagram.png"
response = requests.get(url)

if response.status_code == 200:
    with open("UC1.2_Simple_Sequence_Diagram.png", "wb") as file:
        file.write(response.content)
    print("Download complete.")
else:
    print("Failed to download file. Status code:", response.status_code)
```

Download all files from a list of urls read from a .csv file

```python
import pandas as pd
import os
import requests


# Load the CSV file from the subfolder 'Beispielmodelle'
file_path = 'Beispielmodelle/UMLFiles_List_V2.0.csv'
df_files = pd.read_csv(file_path, sep=',')
print(df_files.head())

# Reformulate the link to be downloadable
df_files['Download Link'] = df_files['Model Link - Github'].str.replace('www.github.com', 'raw.githubusercontent.com').str.replace('/tree/', '/')
failed_downloads = []
countfiles = 0
# Example of reformulated link
print(df_files['Download Link'].head())


# Create directories if they don't exist
directories = ['JPG', 'PNG', 'UML', 'XMI', 'OTHER']
for directory in directories:
    if not os.path.exists(directory):
        os.makedirs(directory)

# Download the first 20 files
for link in df_files['Download Link']:
    if countfiles > 54183:
        file_name = link.split('/')[-1]
        # Determine the subfolder based on file extension
        _, extension = os.path.splitext(file_name)
        extension = extension.lower()
        if extension == '.jpg':
            subfolder = 'JPG'
        elif extension == '.png':
            subfolder = 'PNG'
        elif extension == '.uml':
            subfolder = 'UML'
        elif extension == '.xmi':
            subfolder = 'XMI'
        else:
            subfolder = 'OTHER'
        
        # Check if file already exists in the respective subfolder, if so, change the file name
        file_path = os.path.join(subfolder, file_name)
        if os.path.exists(file_path):
            base, extension = os.path.splitext(file_name)
            counter = 1
            new_file_name = f"{base}_{counter}{extension}"
            new_file_path = os.path.join(subfolder, new_file_name)
            while os.path.exists(new_file_path):
                counter += 1
                new_file_name = f"{base}_{counter}{extension}"
                new_file_path = os.path.join(subfolder, new_file_name)
            file_name = new_file_name
            file_path = new_file_path
        try: 
            response = requests.get(link)
            if response.status_code == 200:
                # Determine the subfolder based on file extension
                _, extension = os.path.splitext(file_name)
                extension = extension.lower()
                if extension == '.jpg':
                    subfolder = 'JPG'
                elif extension == '.png':
                    subfolder = 'PNG'
                elif extension == '.uml':
                    subfolder = 'UML'
                elif extension == '.xmi':
                    subfolder = 'XMI'
                else:
                    subfolder = 'OTHER'
                
                # Save the file in the appropriate subfolder
                file_path = os.path.join(subfolder, file_name)
                try:
                    with open(file_path, "wb") as file:
                        file.write(response.content)
                        print(f"Downloaded: {file_path}")
                except Exception as e:
                    print(f"Failed to write {file_name}. Error:", e)
                    # Add the failed URL to the list
                    failed_downloads.append(link)
            else:
                print(f"Failed to download {file_name}. Status code:", response.status_code)
                # Add the failed URL to the list
                failed_downloads.append(link)
            
        except requests.exceptions.RequestException as e:
            print(f"Failed to download {file_name}. Error:", e)
            # Add the failed URL to the list
            failed_downloads.append(link)
            continue

        # Write the list of failed downloads to a file
        with open("failedDownloads.txt", "w") as f:
            for item in failed_downloads:
                f.write("%s\n" % item)
    countfiles += 1
    print(countfiles)
```
